```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

## 4 LaPlace Jam Sesh

```python
### LaPlace Solution
def laplace(x,t, c): 
    term1 = np.sin(np.pi*x)*np.sin(np.pi*c*t)
    term2 = np.pi*c
    return term1/term2
```

```python
### Seperation of Variables Solution 
def sov(x,t): 
    term1 = ((x**-2)+ 2*t)**-1
    return 1/(1+term1)
```

```python
### We can use the same mesh for each solution: 
#### Parameters & Meshgrid Set-Up 
xl, xr = 0, 5  ### x left and x right
tl, tr = 0, 5  ### t left and t right
c = 1.0        ### some constant, c

#### Set up space for meshgrid with table
xx = np.linspace(xl, xr,50)
tt = np.linspace(tl,tr,50)

## intialize meshgrid (our 3D space)
xxx, ttt = np.meshgrid(xx, tt, sparse=False)
```

```python
## solve for solution(s):
soln_laplace = laplace(xxx, ttt, c)
```

```python
## Finally we can plot: 
### plot in 2D
plt.imshow(soln_laplace,cmap='jet')
plt.xlabel("position (x)")
plt.ylabel("time(t)")
plt.title("LaPlace Jam Sesh: space-time plot")
plt.colorbar() 
```
