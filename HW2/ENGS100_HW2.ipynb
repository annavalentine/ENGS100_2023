## Problem 5: 

```python
def sandy_table(x, y, mu, r):
    sqrt = ((x**2)+(y**2))**(1/2)
    return mu*(sqrt - r)
```

```python
#### Parameters & Meshgrid Set-Up 
r = 1                # table radius (m) 
alpha = 30           # angle (degrees)
mu = np.tan(alpha)   # angle of repose (mu)

#### Set up space for meshgrid with table
xx = np.linspace(-r, r,50)
yy = np.linspace(-r,r,50)

## intialize meshgrid (our 3D space)
xxx, yyy = np.meshgrid(xx, yy, sparse=False)
```

```python
## solve for solution
soln = sandy_table(xxx, yyy, mu, r)
```

```python
### plot in 2D
plt.imshow(soln,cmap='jet')
plt.title("2D Pit of Charpit")
plt.colorbar()
```

```python
### Plot in 3D
ax = plt.axes(projection='3d')
ax.plot_surface(xxx, yyy, soln, cmap="plasma", linewidth=0, antialiased=False, alpha=0.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("3D: Sandcastles r cewl")
# rotation
#ax.view_init(30, 60)

plt.show()
```
